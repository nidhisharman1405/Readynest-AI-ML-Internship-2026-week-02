# Key Performance Indicators

````python
# Key Performance Indicators

# 1. Map columns dynamically to match case variations (e.g., lowercase 'sales')
col_map = {col.lower().replace("_", " "): col for col in retail_df.columns}

sales_col = col_map.get('sales')
profit_col = col_map.get('profit')
cust_col = col_map.get('customer id') or col_map.get('customerid')
order_col = col_map.get('order id') or col_map.get('orderid')

# 2. Run your exact calculations using the mapped columns
sales = retail_df[sales_col].sum() if sales_col else 0
profit = retail_df[profit_col].sum() if profit_col else 0
customers = retail_df[cust_col].nunique() if cust_col else 0
orders = retail_df[order_col].nunique() if order_col else 0
avg_order = sales / orders if orders > 0 else 0

# 3. Your exact print statements
print("Total Sales :", sales)
print("Total Profit :", profit)
print("Customers :", customers)
print("Orders :", orders)
print("Average Order Value :", avg_order)
````
