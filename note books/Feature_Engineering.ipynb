# Business Feature Construction

```python
import pandas as pd

retail_df=pd.read_csv('../data/processed/retail_cleaned.csv')
```

```python
retail_df['Order Date']=pd.to_datetime(retail_df['Order Date'])
```

```python
retail_df['Ship Date']=pd.to_datetime(retail_df['Ship Date'])
```

```python
retail_df['Date of Birth']=pd.to_datetime(retail_df['Date of Birth'])
```

```python
retail_df['Month']=retail_df['Order Date'].dt.month_name()
```

```python
retail_df['Year']=retail_df['Order Date'].dt.year
```

```python
retail_df['Quarter']=retail_df['Order Date'].dt.quarter
```

```python
retail_df['Weekday']=retail_df['Order Date'].dt.day_name()
```

```python
retail_df['Age']=2026-retail_df['Date of Birth'].dt.year
```

```python
retail_df['Delivery Days']=(
retail_df['Ship Date']
-
retail_df['Order Date']
).dt.days
```

```python
retail_df['Profit Margin']=(
retail_df['Profit']/retail_df['Sales']
)*100
```
