# Retail Performance Exploration

```python
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
import pandas as pd
retail_df=pd.read_csv('../data/processed/retail_cleaned.csv')
```

```python
#Sales Distribution

sns.histplot(retail_df['Sales'])
```

```python
#Profit Distribution

sns.histplot(retail_df['Profit'])
```

```python
#Age Distribution

sns.histplot(retail_df['Age'])
```

```python
#Region

sns.countplot(

data=retail_df,

x='Region'

)
```

```python
#City Type

sns.countplot(

data=retail_df,

x='City Type'

)
```

```python
#Outlet Type

sns.countplot(

data=retail_df,

x='Outlet Type'

)
```

```python
(
    retail_df.groupby('Category of Goods')['Sales']
    # Category
    .sum()
    .plot(kind='bar')
)
```

```python
(
    retail_df.groupby('Sub-Category')['Sales']
    # Subcategory
    .sum()
    .nlargest(10)
    .plot(kind='bar')
)
```

```python
# Top customers
cust = (
    retail_df.groupby('Customer Name')['Sales']
    .sum()
    .nlargest(10)
)

cust.plot(kind='barh')
```

```python
# Top Products
prod = (
    retail_df.groupby(
        'Product Name'
    )['Sales']
    .sum()
    .nlargest(10)
)

prod.plot(
    kind='barh'
)
```

```python
# Monthly Trend
monthly = (
    retail_df.groupby(
        'Month'
    )['Sales']
    .sum()
)

monthly.plot()
```

```python
# Quarterly
quarter = (
    retail_df.groupby(
        'Quarter'
    )['Sales']
    .sum()
)

quarter.plot(
    kind='bar'
)
```

```python
# Correlation Heatmap

sns.heatmap(

retail_df.select_dtypes(

include='number'

).corr(),

annot=True

)
```

```python
# Delivery Days

sns.histplot(

retail_df['Delivery Days']
)
```

```python
# Profit Margin

sns.histplot(

retail_df['Profit Margin']
)
```
